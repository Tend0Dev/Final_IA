* Nodos: Representan condicionales o pruebas sobre las caracter√≠sticas de los datos
      (¬øEl monto de la transacci√≥n es mayor a    $10,000?).

* Ramas: Son las posibles respuestas a esas condiciones
      ("Verdadero" o "Falso").

* Hojas: Son los resultados finales o clases
      ("Fraudulento" o "No fraudulento").

## Paso 1: Limpieza y Transformaci√≥n de Coordenadas

## Explicaci√≥n Paso 1

### Problema
Las coordenadas estaban almacenadas en un formato incorrecto, por ejemplo: `-112.154.481`, lo que causaba errores al intentar convertirlas a valores num√©ricos (`float`).

---

### Soluci√≥n

Se implementa una funci√≥n `clean_coordinate` para detectar y corregir los siguientes casos:

1. **Coordenadas con m√∫ltiples puntos decimales**  
   - Ejemplo: `-112.154.481` se convierte en `-112154481` y luego se transforma a float dividi√©ndolo por una potencia de 10.

2. **Coordenadas con comas como separador decimal**  
   - Ejemplo: `-112,154` ‚Üí `-112.154`

3. **Valores ya v√°lidos**  
   - Se dejan tal cual si no requieren correcci√≥n.

---

### Validaci√≥n

Despu√©s de la limpieza, se validan las coordenadas geogr√°ficas:

-  **Latitud** debe estar entre `-90` y `90`  
-  **Longitud** debe estar entre `-180` y `180`  

Las coordenadas que no cumplan con estos rangos son descartadas.

---

### Resultado

Se obtiene un conjunto de coordenadas limpias, en formato `float`, listas para ser utilizadas en c√°lculos como distancias geogr√°ficas, clustering, o visualizaci√≥n en mapas.


```python
def clean_coordinate(coord):
    """
    Versi√≥n mejorada con depuraci√≥n para identificar problemas
    """
    try:
        # Si es string, limpiarlo primero
        if isinstance(coord, str):
            coord = coord.strip()  # Eliminar espacios en blanco

            # Caso 1: M√∫ltiples puntos (ej: -112.154.481)
            if coord.count('.') > 1:
                parts = coord.split('.')
                # Verificar que todas las partes son d√≠gitos
                if not all(part.lstrip('-').isdigit() for part in parts):
                    raise ValueError(f"Partes no num√©ricas en coordenada: {coord}")
                whole_num = parts[0] + ''.join(parts[1:])
                scale = 10 ** (len(parts)-1)
                return float(whole_num) / scale

            # Caso 2: Coma como separador decimal
            if ',' in coord:
                return float(coord.replace(',', '.'))

            # Caso 3: Formato normal
            return float(coord)

        # Si ya es num√©rico
        return float(coord)

    except Exception as e:
        print(f"‚ö†Ô∏è Error procesando coordenada '{coord}': {str(e)}")
        return np.nan

# Aplicar con verificaci√≥n
print("\nüîç Antes de limpieza:")
print(df[coord_cols].head())

df[coord_cols] = df[coord_cols].applymap(clean_coordinate)

print("\nüîç Despu√©s de limpieza:")
print(df[coord_cols].head())

# Filtrar solo filas con coordenadas v√°lidas
valid_coords = df[coord_cols].notna().all(axis=1)
df = df[valid_coords].copy()

print("\nEstad√≠sticas finales:")
print(df[coord_cols].describe())
```

## Paso 2: C√°lculo de Distancias Geogr√°ficas

### Objetivo
Calcular la distancia entre la ubicaci√≥n del cliente y el comercio con el fin de detectar **transacciones imposibles geogr√°ficamente**, es decir, compras en lugares muy alejados en poco tiempo.

---

### Implementaci√≥n

1. ** F√≥rmula de Haversine**  
   Se utiliza para calcular la distancia entre dos puntos geogr√°ficos sobre la superficie terrestre, teniendo en cuenta la curvatura de la Tierra.

---

###  Protecciones y Validaciones

-  **Conversi√≥n expl√≠cita a `float`**  
  Asegura que los datos de coordenadas est√©n en el tipo correcto antes del c√°lculo.

-  **Conversi√≥n a radianes**  
  Requisito para funciones trigonom√©tricas (`sin`, `cos`, `acos`) en la f√≥rmula de Haversine.

- **Bloque `try-except`**  
  Captura posibles errores durante la conversi√≥n o c√°lculo, como valores nulos o no num√©ricos.

-  **Radio de la Tierra**  
  Se usa un radio de `6371 km` para obtener el resultado en kil√≥metros.

---

### Resultado
Se obtiene una columna nueva con la **distancia geogr√°fica** entre cliente y comercio, √∫til como variable para detectar posibles fraudes o inconsistencias.


```python
def safe_haversine(row):
    """Calcula distancia en km entre coordenadas con validaci√≥n robusta"""
    try:
        lat1, lon1 = radians(float(row['lat'])), radians(float(row['long']))
        lat2, lon2 = radians(float(row['merch_lat'])), radians(float(row['merch_long']))

        dlat = lat2 - lat1
        dlon = lon2 - lon1
        a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
        return 6371 * 2 * atan2(sqrt(a), sqrt(1-a))  # 6371 = radio terrestre en km
    except:
        return np.nan

df['distancia_compra'] = df.apply(safe_haversine, axis=1)
```

## Paso 4: Variables de Agregaci√≥n por Usuario

### Objetivo
Crear caracter√≠sticas que capturen el comportamiento hist√≥rico del usuario, lo que ayuda a detectar patrones an√≥malos que puedan indicar fraude.

---

### Variables clave

- ** `historial_24h`**  
  N√∫mero de transacciones realizadas por el usuario en las √∫ltimas 24 horas.  
  - **Prop√≥sito:** Detectar **r√°fagas sospechosas** de actividad en un corto per√≠odo de tiempo, lo que puede ser un indicio de fraude.

- ** `avg_amt_category`**  
  Promedio hist√≥rico de gasto por categor√≠a de transacciones del usuario.  
  - **Prop√≥sito:** Identificar **gastos an√≥malos** en comparaci√≥n con los patrones hist√≥ricos del usuario, lo que puede se√±alar comportamientos fraudulentos.

---

### T√©cnicas

- ** `rolling('24h')`**  
  Usa una **ventana temporal m√≥vil** de 24 horas para calcular el n√∫mero de transacciones en ese per√≠odo, lo que ayuda a detectar cambios repentinos en la actividad.

- ** `expanding().mean()`**  
  Calcula el **promedio acumulativo** de los gastos de un usuario, permitiendo detectar desviaciones a lo largo del tiempo en las diferentes categor√≠as de gasto.

---

###  Beneficio
Estas variables ayudan a identificar patrones de comportamiento hist√≥rico que pueden indicar transacciones sospechosas, permitiendo al modelo detectar fraudes de manera m√°s efectiva al analizar el comportamiento pasado de cada usuario.


```python
# PASO 1: Versi√≥n robusta del c√°lculo de agregaciones
try:
    # Asegurar que las columnas cr√≠ticas existen
    required_cols = ['cc_num', 'trans_date_trans_time', 'amt']
    missing_cols = [col for col in required_cols if col not in df.columns]

    if missing_cols:
        raise KeyError(f"Columnas faltantes: {missing_cols}")

    # Conversi√≥n segura de fecha
    if not pd.api.types.is_datetime64_any_dtype(df['trans_date_trans_time']):
        df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'], errors='coerce')

    # Eliminar filas con fechas inv√°lidas (si las hay)
    df = df.dropna(subset=['trans_date_trans_time'])

    # Ordenamiento seguro
    df = df.sort_values(['cc_num', 'trans_date_trans_time'])

    # C√°lculo robusto de historial 24h
    df['historial_24h'] = (
        df.groupby('cc_num', group_keys=False)
        .apply(lambda x: x.set_index('trans_date_trans_time')['amt']
        .rolling('24h')
        .count())
        .reset_index(level=0, drop=True)
    )

    # C√°lculo robusto de promedio por categor√≠a
    if 'category' in df.columns:
        df['avg_amt_category'] = (
            df.groupby(['cc_num', 'category'], group_keys=False)['amt']
            .expanding()
            .mean()
            .reset_index(level=[0,1], drop=True)
        )
    else:
        df['avg_amt_category'] = (
            df.groupby('cc_num', group_keys=False)['amt']
            .expanding()
            .mean()
            .reset_index(level=0, drop=True)
        )

    # Manejo de valores nulos
    df['historial_24h'] = df['historial_24h'].fillna(0)
    df['avg_amt_category'] = df['avg_amt_category'].fillna(0)

    print("\n‚úÖ Variables calculadas exitosamente:")
    print(df[['cc_num', 'trans_date_trans_time', 'amt', 'historial_24h', 'avg_amt_category']].head(3))

except Exception as e:
    print(f"\n‚ùå Error durante el c√°lculo: {str(e)}")
```

   Antes de intentar codificar la columna `merchant`, se revisa si realmente existe en el DataFrame.

2. **Evitar duplicados**

---

###  Mensajes Claros

- Se muestran mensajes explicativos sobre las operaciones realizadas, lo que permite saber si la columna `category` fue encontrada o si se realizaron otras codificaciones.
- En caso de que falten las columnas esperadas, se advierte de manera clara sobre la ausencia de estas.

###  Beneficio
La codificaci√≥n adecuada de las variables categ√≥ricas mejora la calidad de los modelos, permiti√©ndoles interpretar caracter√≠sticas como categor√≠as de transacciones o comercios de manera efectiva.

## Paso 6: Entrenamiento del Modelo con Pipeline Robusto

### Objetivo
Entrenar un modelo de **Random Forest** utilizando un pipeline robusto para manejar valores faltantes, desbalance de clases y evitar sobreajuste.

---

### Arquitectura del Pipeline

1. **Imputaci√≥n:**
   - Se utilizan t√©cnicas de imputaci√≥n para rellenar cualquier valor faltante en las caracter√≠sticas num√©ricas con la mediana.
   - Esto proporciona una protecci√≥n adicional, evitando que los valores faltantes afecten negativamente el rendimiento del modelo.

2. **SMOTE:**
   - Se emplea el **SMOTE (Synthetic Minority Over-sampling Technique)** para generar muestras sint√©ticas de la clase minoritaria (fraudes).
   - Esto ayuda a balancear las clases en el conjunto de datos, mejorando la capacidad del modelo para aprender sobre fraudes.

3. **Random Forest:**
   - **`class_weight='balanced_subsample'`**: Ajusta los pesos de las clases para manejar el desbalance de clases, d√°ndole m√°s importancia a la clase minoritaria (fraude).
   - **`max_depth=12`**: Controla la profundidad m√°xima de los √°rboles para evitar el sobreajuste.
   - **`n_jobs=-1`**: Utiliza todos los n√∫cleos del procesador disponibles para paralelizar el entrenamiento, acelerando el proceso.

---

### C√°lculo Condicional de Variables

1. **Verificaci√≥n de las columnas de agregaci√≥n:**
   - Se verifica si las columnas de agregaci√≥n (`historial_24h`, `avg_amt_category`) existen en el DataFrame.
   - Si no est√°n presentes, se calculan utilizando diferentes enfoques seg√∫n las columnas disponibles.

2. **Manejo de casos especiales:**
   - Si no existe la columna `cc_num` (n√∫mero de tarjeta), se crea un ID temporal (`temp_id`) combinando otras columnas, como la ciudad y el trabajo del usuario.
   - Se emplean distintas estrategias para calcular promedios, dependiendo de los datos disponibles (por ejemplo, `category` o `cat_*`).

---

### Selecci√≥n Robusta de Features

1. **Construcci√≥n de la lista de caracter√≠sticas (features):**
   - Se verifica la existencia de cada columna antes de incluirla en el conjunto de caracter√≠sticas para el modelo.
   - Se proporcionan mensajes claros si faltan columnas esperadas, asegurando que el modelo solo reciba datos v√°lidos.

---


### Beneficio
Este enfoque robusto garantiza que el modelo est√© bien entrenado, manejando datos desbalanceados, valores faltantes y diferentes configuraciones de las caracter√≠sticas, lo que mejora su capacidad para detectar fraudes de manera precisa.


```python
# (Evita recalcular si ya fueron generadas antes)
if 'historial_24h' not in df.columns or 'avg_amt_category' not in df.columns:

    # Asegurarse de tener una columna de identificaci√≥n de usuario
    if 'cc_num' not in df.columns and 'trans_date_trans_time' not in df.columns:
        # Si no hay columnas clave como 'cc_num' o 'trans_date_trans_time', se crea un ID temporal combinando ciudad y trabajo
        df['temp_id'] = df['city_encoded'].astype(str) + "_" + df['job'].astype(str)
        group_col = 'temp_id'
    else:
        # Usar 'cc_num' si est√° disponible; si no, usar 'trans_num' como identificador
        group_col = 'cc_num' if 'cc_num' in df.columns else 'trans_num'

    # Historial de transacciones: se calcula cu√°ntas transacciones ha hecho cada usuario
    df['historial_24h'] = df.groupby(group_col)['amt'].transform('count')

    # Promedio del monto por categor√≠a de compra
    if 'category' in df.columns:
        # Si hay una columna 'category', se usa directamente para agrupar
        df['avg_amt_category'] = df.groupby([group_col, 'category'])['amt'].transform('mean')
    else:
        # Si no hay 'category', se buscan columnas one-hot como 'cat_food', 'cat_gas', etc.
        cat_cols = [col for col in df.columns if col.startswith('cat_')]
        if cat_cols:
            # Se identifica la categor√≠a activa por fila (la que tiene el valor m√°s alto)
            df['main_cat'] = df[cat_cols].idxmax(axis=1)
            df['avg_amt_category'] = df.groupby([group_col, 'main_cat'])['amt'].transform('mean')
        else:
            # Si no hay categor√≠as, se usa el promedio por grupo general
            df['avg_amt_category'] = df.groupby(group_col)['amt'].transform('mean')

# Paso 2: Selecci√≥n robusta de caracter√≠sticas
# Se arma una lista final de variables a usar como entrada del modelo
available_features = []
for feature in ['amt', 'distancia_compra', 'hora_dia', 'dia_semana',
                'historial_24h', 'avg_amt_category']:
    if feature in df.columns:
        available_features.append(feature)
    else:
        print(f" Advertencia: Columna '{feature}' no encontrada")  # Mensaje de advertencia si falta una columna clave

# A√±adir columnas categ√≥ricas (one-hot) y otras relevantes codificadas
cat_features = [col for col in df.columns if col.startswith('cat_')]
other_features = ['merchant_encoded', 'city_encoded']  # Codificaciones adicionales

# Combinar todo en la lista final de features
features = available_features + cat_features + [col for col in other_features if col in df.columns]


# Paso 3: Crear los conjuntos de datos
X = df[features]       # Variables independientes
y = df['is_fraud']     # Variable objetivo (1 = fraude, 0 = normal)

# Paso 4: Pipeline de modelado con limpieza y balanceo
pipeline = make_pipeline(
    SimpleImputer(strategy='median'),  # Imputaci√≥n de valores faltantes con la mediana
    SMOTE(random_state=42),            # Balanceo de clases usando sobremuestreo sint√©tico
    RandomForestClassifier(
        n_estimators=250,              # N√∫mero de √°rboles
        max_depth=12,                  # Profundidad m√°xima del √°rbol
        class_weight='balanced_subsample',  # Ajuste de peso para manejar clases desbalanceadas
        random_state=42,               # Reproducibilidad
        n_jobs=-1                      # Uso de todos los n√∫cleos disponibles
    )
)

# Divisi√≥n de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar el modelo con el pipeline definido
pipeline.fit(X_train, y_train)
print("\n Modelo entrenado correctamente")
```

## Paso 7: Evaluaci√≥n y M√©tricas

### Objetivo
Evaluar el rendimiento del modelo utilizando varias m√©tricas clave para entender su capacidad de detectar fraudes, especialmente en un contexto de datos desbalanceados.

---

###  M√©tricas Clave

1. **Accuracy (Precisi√≥n General):**
   - Representa el porcentaje de predicciones correctas realizadas por el modelo.
   - **Advertencia:** En un conjunto de datos desbalanceado (como fraudes), esta m√©trica puede ser enga√±osa. Un modelo que siempre predice la clase mayoritaria puede tener un alto accuracy, pero no estar detectando correctamente los fraudes.

2. **AUC-ROC (√Årea bajo la curva ROC):**
   - Mide la capacidad del modelo para distinguir entre las clases (fraude y no fraude).
   - Idealmente, el valor deber√≠a ser mayor a 0.9 para indicar una buena discriminaci√≥n entre las clases.
   - **Interpretaci√≥n:** Un AUC cercano a 1 sugiere un excelente rendimiento del modelo para separar fraudes de transacciones leg√≠timas.

---

###  Reporte de Clasificaci√≥n

1. **Precision (Precisi√≥n):**
   - Mide el porcentaje de verdaderos fraudes entre los casos que el modelo ha predicho como fraude.
   - F√≥rmula: `Precision = TP / (TP + FP)`  
     Donde `TP` son los verdaderos positivos y `FP` los falsos positivos.

2. **Recall (Sensibilidad o Exhaustividad):**
   - Mide el porcentaje de fraudes reales que el modelo ha detectado correctamente.
   - F√≥rmula: `Recall = TP / (TP + FN)`  
     Donde `TP` son los verdaderos positivos y `FN` los falsos negativos.

3. **F1-Score:**
   - Es una medida combinada que balancea la precisi√≥n y el recall.
   - F√≥rmula: `F1-Score = 2 * (Precision * Recall) / (Precision + Recall)`
   - Es especialmente √∫til cuando se busca un balance entre precisi√≥n y recall en tareas de clasificaci√≥n con clases desbalanceadas.

---

###  Matriz de Confusi√≥n

1. **Visualizaci√≥n de errores:**
   - La matriz de confusi√≥n muestra el n√∫mero de **verdaderos positivos (TP)**, **falsos positivos (FP)**, **verdaderos negativos (TN)** y **falsos negativos (FN)**.
   - Es crucial para **entender los errores** del modelo, especialmente los falsos positivos y falsos negativos.
   - En el contexto de fraude, minimizar los **falsos negativos** (fraudes no detectados) es m√°s importante, pero tambi√©n es necesario controlar los **falsos positivos** (leg√≠timos clasificados como fraude).

---

### Beneficio
Estas m√©tricas permiten una evaluaci√≥n detallada del modelo, ayudando a identificar posibles √°reas de mejora en la detecci√≥n de fraudes y asegurando que el modelo sea eficiente en su tarea sin estar sesgado hacia una clase mayoritaria.


```python
# Predicciones
y_pred = pipeline.predict(X_test)
y_proba = pipeline.predict_proba(X_test)[:, 1]

# M√©tricas
print("\nüìà M√©tricas de evaluaci√≥n:")
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print(f"AUC-ROC: {roc_auc_score(y_test, y_proba):.4f}")
print("\nReporte de Clasificaci√≥n:")
print(classification_report(y_test, y_pred))

# Matriz de Confusi√≥n
print("\nüéØ Matriz de Confusi√≥n:")
ConfusionMatrixDisplay.from_estimator(pipeline, X_test, y_test,
                                    display_labels=['Leg√≠tima', 'Fraude'],
                                    cmap='Blues')
plt.show()
```
